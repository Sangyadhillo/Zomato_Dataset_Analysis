# Data Cleaning

## These columns are not important so dropping them

```python
df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1,inplace=True)
df
```

## cleaning Rate Column

```python
df.rate.unique()
```

##  Removing 'NEW',-' and '/5' from the rate column

```python
def rateremove(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(rateremove)
df.head()
```

## checking  null values in rate column


```python
df.rate.isnull().sum()
```

```python
# Filling Null values in rate column by mean
df['rate'].fillna(round(df.rate.mean(),2),inplace=True)
```

```python


df.info()
```

##  checking percentage of rest of the null values

```python
df.isnull().sum()
```

```python
percentage=round((df.isnull().sum()/df.shape[0])*100,2)
```

```python
percentage
```

```python
# dropping Null values
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

##  renaming the big columns 

```python
df.columns
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace =True)
df.head()
```


##  checking columns city and location have same values

```python
df.location.unique()
```

```python
df.city.unique()
```

##  we will drop city column as it is same like location

```python
df.drop('city',axis=1,inplace=True)
df.head()
```

## checking the data type of cost2plates column

```python
df.info()
```

```python
df.cost2plates.unique()
```


## Removing , from cost2plates column

```python



def makestr(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost2plates']=df['cost2plates'].apply(makestr)
df.cost2plates.unique()
```

```python
df.head()
```



# Cleaning rest_type column

```python
df.rest_type.value_counts()
```

# Making values which are less tahn 100 into one group  that is other

```python
rest_type=df['rest_type'].value_counts()
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

# Making the values of rest_type_lessthan1000 as others in our DataFrame

```python
def handle_rest_type(value):
    if value in rest_type_lessthan1000:
        return 'Others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

# Cleaning Location column

```python
df.location.value_counts()
```

```python
# Making a cluster of those location which have less restraunts
```

```python
location=df['location'].value_counts()
location
```

```python
location_lestthan200=location[location<200]
location_lestthan200.head()
```

```python
# making a function to name location_lestthan200 these value as others
```

```python
def handle_location(value):
    if value in location_lestthan200:
        return 'Others'
    else:
        return value
df['location']=df['location'].apply(handle_location)
df.location.value_counts().head(10)
```

```python
df.head()
```

# Cleaning cuisines column 

```python
df.cuisines.value_counts()
```

```python
# create a function for cuisines which are less than 100 we will gonna put them under others category 
```

```python
cuisines=df['cuisines'].value_counts()
cuisines_lessthan100=cuisines[cuisines<100]
cuisines_lessthan100

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'Others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```
